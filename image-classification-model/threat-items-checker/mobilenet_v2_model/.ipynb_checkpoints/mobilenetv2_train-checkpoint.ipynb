```python
X, y = [], []
for image_path in image_list:
    image = cv2.imread(str(image_path))
    if image is None:
        print(f"Skipping unreadable image: {image_path}")
        continue

    label_file = labels_dir / (image_path.stem + ".txt")
    if label_file.exists():
        label_content = label_file.read_text().strip()
        if label_content:
            label_data = label_content.split()
            if len(label_data) > 0:
                label = int(label_data[0])
                resized_img = cv2.resize(image, IMAGE_SHAPE)
                X.append(resized_img)
                y.append(label)
            else:
                print(f"Warning: No valid label in {label_file}. Removing file.")
                label_file.unlink()
        else:
            print(f"Warning: Empty label file: {label_file}. Removing file and image.")
            label_file.unlink()
            image_path.unlink()
    else:
        print(f"Warning: No label found for {image_path}")

```

```python
X[0]/255
```

```python
X=np.array(X)
y=np.array(y)
X.shape,y.shape
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,random_state=0)
```

```python
import tensorflow as tf
import tensorflow_hub as hub
import tf_keras
IMAGE_SHAPE = (224, 224)

feature_extractor_model = "https://tfhub.dev/google/tf2-preview/mobilenet_v2/feature_vector/4"
pretrained_model_without_top_layer = hub.KerasLayer(
    feature_extractor_model, input_shape=IMAGE_SHAPE + (3,), trainable=False
)

num_of_weapons = 5

model = tf_keras.Sequential([
    pretrained_model_without_top_layer,
    tf_keras.layers.Dense(num_of_weapons)
])

model.summary()

model.compile(
    optimizer="adam",
    loss=tf.keras.losses.SparseCategoricalCrossentropy(from_logits=True),
    metrics=["accuracy"]
)

model.fit(X_train_scaled, y_train, epochs=5)
```
